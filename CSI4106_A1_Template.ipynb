```python

high_risk_data = dataset[dataset['RiskLevel'] == 'high risk']  #get all records that have high risk level


BodyTemp_highRisk = high_risk_data.groupby("Age").agg({'SystolicBP': 'mean'})  #Get the average SystolicBP for every age that has high risk

# Print the result
print(BodyTemp_highRisk)
```

For younger age groups (like age 12, 13, 14), the systolic blood pressure is around 90 mmHg.
As we move to older age groups (like 30, 35, 40, etc.), systolic blood pressure rises, with some values reaching as high as 140 mmHg.
at certain ages (e.g., 29, 40, 42), we observe average systolic blood pressure around 135 mmHg, which could be a warning sign for elevated cardiovascular risk. Something to note, is that certain ages like 15,16, and 20 don't have a single record individual that has high risk.
